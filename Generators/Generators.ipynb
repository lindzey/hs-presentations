```python
# TODO: is this the best way to handle constants like this?
# NOPE!! For this description, I probably want to go ahead and use a namedtuple. 
T_IDX = 0
Z_IDX = 3
L_IDX = 5 # label; 1 for surface, 0 for noise

# appends a 6th field to the photon; if it is flipped to 0, that indicates 
# that the photon has been labeled as some form of noise
def read_line(line):
    ts, x, y, z, ch = line.split()
    return [float(ts), float(x), float(y), float(z), int(ch), 1]
```

Now, we need to find a representation for the data so that it's easy to calculate the local density for a single photon. I do this by grouping the return points into pulses (all returns from a given laser pulse) and windows (set of subsequent pulses). This is easy to do because our input data is already ordered in time. 

However, the data files we work with have millions or billions of points, we are unable to do everything in memory. So, we first need a way to work with only a subset of the data at a time. Python generators are a natural fit for this, and free me from having to worry about maintaining the correct buffers of data.

```python
def stream_photons(filename):
    for line in open(filename):
        yield read_line(line)
```

Now, we have a generator that yields individual photons from the file. The 'yield' keyword indicates that a function is a generator. After a generator has been created, each subsequent call of 'next' causes the function to execute, starting from wherever it last yielded, until it reaches a yield statement again, at which point it returns that value. 

For each laser pulse we will receive a (potentially large) number of photons, since we have 100 detectors, each capable of detecting multiple photons per pulse.

We want to group the photons by timestamp in order to reason about groups based on which pulse they belong to. We do this using another generator! Notes:

* there is no way to know ahead of time how many photons will belong in this group; the for loop runs until a photon is found that doesn't match the timestamp, at which point the existing pulse is returned via 'yield'. Upon the next call, the photon that didn't match is added to the group, and the process continues. 

* we can concatenate the generators - and the chain will automatically handle reading in the correct amount of data!

* At no point in this process do we care how the file was loaded - both the map() implementation and the generator implementation will provide an iterable object that works in all these cases! 

```python
# This yields pulses, which are a list of all photons with the same timestamp
# Example Usage:
# pulses = stream_pulses(stream_photons('ICP4.F23.0d.mcmurdo1.txyzs'))
def stream_pulses(photon_stream):
    group = []
    for photon in photon_stream:
        if group != [] and photon[T_IDX] != group[0][T_IDX]:
            yield group
            group = []
        group.append(photon)
```

Finally, we pull the data into a "window" data structure.

* centered at a given pulse

* contains all the photons "close" to those in the pulse one we're trying to label

* automatically advances by a single pulse to return overlapping data with each call 

```python
# This yields overlapping windows that advance by one pulse at a time
# and contain 2*window_radius+1 pulses
def stream_windows(pulse_stream, window_radius):
    window = []
    for pulse in pulse_stream:
        if len(window) < 2*window_radius + 1:
            window.append(pulse)
        else:  # only want to run this if we have a full swath
            yield window 
            window = window[1:]   # remove first pulse
            window.append(pulse) # add new pulse to the end    
```

* As expected, we see pulses of photons grouped in time, and can see a clear surface, even with only 40 pulses ~= 2ms of data.

* In this plot, darker points indicate photons densely grouped in time and space on top of each other. 

* The presence of temporally-isolated high-density areas indicates that we need to look for areas of high density that are also next to other high-density areas; so, we'll filter based on how many photons from nearby pulses are close to the input photon, ignoring photons from the same pulse.

```python
# This yields one pulse worth of photons, annotated with the local 
# density (number of other photons in the pulse that are within
# neighbor_thresh of their z)
# If density_threshold is set, also require that the last field of the
# photon is greater than the density_threshold
def filter_solar(window_stream, neighbor_threshold, density_threshold):
    for window in window_stream:
        l = len(window)
        center_idx = (l-1)/2
        center_pulse = window[center_idx]
        other_pulses = [pulse for idx,pulse in enumerate(window) if idx!=center_idx]
        other_photons = [photon for sublist in other_pulses for photon in sublist]
        # TODO: make my own count function to make this more readable ... (possibly count = sum)
        for photon in center_pulse:
            density = len([1 for other in other_photons if 
                           abs(photon[Z_IDX]-other[Z_IDX]) < neighbor_threshold and 
                           other[L_IDX] != 0])
            if density < density_threshold:
                photon[L_IDX] = 0
        yield center_pulse
```

In the output of this filter, the 6th field indicates whether the photon has been labeled signal or noise. To check its performance, let's plot all the photons that are labeled 'surface':
