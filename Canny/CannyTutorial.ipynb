```python
# from http://stackoverflow.com/questions/12201577/convert-rgb-image-to-grayscale-in-python
def rgb2gray(rgb):
    return np.dot(rgb[...,:3], [0.299, 0.587, 0.144])

im_gray = rgb2gray(im)    
plt.imshow(im_gray, cmap = plt.get_cmap('gray'))
plt.show()
```

```python
# returns a blurred version of the image, where the kernel is 2*radius+1 wide 
# (kernel needs to have odd dimensions)
def blur_filter(im, radius):
    width = 2*radius + 1
    kernel = np.ones(width*width).reshape(width,width)
    new_im = scipy.ndimage.convolve(im, kernel)
    return new_im
```

```python
figsize(12,9)
subplot(2,2,1)
imshow(im_gray, cmap=plt.get_cmap('gray'))
title('Original Image')
subplot(2,2,3)
imshow(blur_filter(im_gray, 0), cmap=plt.get_cmap('gray'))
title('Kernel width 1 - same as orig image')
subplot(2,2,2)
imshow(blur_filter(im_gray, 2), cmap=plt.get_cmap('gray'))
title('Kernel width 5')
subplot(2,2,4)
imshow(blur_filter(im_gray,4), cmap=plt.get_cmap('gray'))
title('Kernel width 9')
```

```python
def generate_gaussian_kernel(radius, sigma):
    width = 2*radius + 1
    kernel = np.zeros(width*width).reshape(width, width)
    A = 1.0/(sigma * sqrt(2*pi))
    for row in xrange(width):
        # calculating distances from center
        d_row = row-radius
        for col in xrange(width):
            d_col = col-radius
            kernel[row][col] = A*exp(-(d_row**2 + d_col**2)/(2*sigma**2))
    return kernel
```

```python
figsize(18,9)
subplot(2,4,1)
k1 = generate_gaussian_kernel(5, 0.5)
imshow(k1, cmap=plt.get_cmap('gray'), interpolation='nearest')
title('Gaussian kernel, radius 5, sigma 0.5')

subplot(2,4,2)
k2 = generate_gaussian_kernel(5, 1.0)
imshow(k2, cmap=plt.get_cmap('gray'), interpolation='nearest')
title('Gaussian kernel, radius 5, sigma 1')

subplot(2,4,3)
k3 = generate_gaussian_kernel(5, 2.0)
imshow(k3, cmap=plt.get_cmap('gray'), interpolation='nearest')
title('Gaussian kernel, radius 5, sigma 2.0')

subplot(2,4,4)
k4 = generate_gaussian_kernel(5, 5.0)
imshow(k4, cmap=plt.get_cmap('gray'), interpolation='nearest')
title('Gaussian kernel, radius 5, sigma 5.0')
```

```python
def blur_gaussian(im, radius, sigma):
    kernel = generate_gaussian_kernel(radius, sigma)
    new_im = scipy.ndimage.convolve(im, kernel)
    return new_im
```

```python
figsize(12,9)
subplot(2,2,1)
imshow(im_gray, cmap=plt.get_cmap('gray'))
title('Original Image')
subplot(2,2,3)
imshow(blur_gaussian(im_gray, 5, 1), cmap=plt.get_cmap('gray'))
title('Gaussian blur: radius = 5, sigma = 1')

subplot(2,2,2)
imshow(blur_gaussian(im_gray, 5, 2), cmap=plt.get_cmap('gray'))
title('Gaussian blur: radius = 5, sigma = 2')
subplot(2,2,4)==]/
imshow(blur_gaussian(im_gray, 5, 5), cmap=plt.get_cmap('gray'))
title('Gaussian blur: radius = 5, sigma = 5')

```
